```python
# Importing Necessary Libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/ipl.csv')
```

```python
df.head()
```

```python
# -----Data Cleaning-----
# Removing unwanted columns
columns_to_remove = ['mid', 'venue', 'batsman', 'bowler', 'striker', 'non-striker']
df.drop(labels = columns_to_remove, axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['bat_team'].unique()
```

```python
# Keeping only consistents teams
consistent_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals', 'Mumbai Indians',
                    'Kings XI Punjab', 'Royal Challengers Bangalore', 'Delhi Daredevils', 'Sunrisers Hyderabad']
```

```python
df = df[(df['bat_team'].isin(consistent_teams)) & (df['bowl_team'].isin(consistent_teams))]
```

```python
# Removing the first 5 overs data in every match: there should be atleast 5 over scores
df = df[df['overs'] >= 5.0]
```

```python
df.head()
```

```python
print(df['bat_team'].unique())
print(df['bowl_team'].unique())
```

```python
# Converting the column 'date' from string into datetime object
from datetime import datetime
df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
df.dtypes
```

```python
# Dealing with Categorical features using the get_dummies method (bat_team and bowl_team columns are categorical columns)
encoded_df = pd.get_dummies(data=df, columns=['bat_team', 'bowl_team'])
```

```python
encoded_df.head()
```

```python
encoded_df.columns
```

```python
encoded_df = encoded_df[['date', 'bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils', 'bat_team_Kings XI Punjab',
                         'bat_team_Kolkata Knight Riders', 'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals', 'bat_team_Royal Challengers Bangalore',
                         'bat_team_Sunrisers Hyderabad', 'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils', 'bowl_team_Kings XI Punjab',
                         'bowl_team_Kolkata Knight Riders', 'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals', 'bowl_team_Royal Challengers Bangalore',
                         'bowl_team_Sunrisers Hyderabad', 'runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5', 'total']]
```

```python
# Splitting the data into train and test set
X_train = encoded_df.drop(labels ='total', axis=1)[encoded_df['date'].dt.year <= 2016]
X_test = encoded_df.drop(labels='total', axis=1)[encoded_df['date'].dt.year >= 2017]
```

```python
y_train = encoded_df[encoded_df['date'].dt.year <= 2016]['total'].values
y_test = encoded_df[encoded_df['date'].dt.year >= 2017]['total'].values
```

```python
# Removing the date column
X_train.drop(labels=['date'], axis=1, inplace=True)
X_test.drop(labels=['date'], axis=1, inplace=True)
```

##**Model building**

```python
# Linear Regressiobn
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()
Lr.fit(X_train, y_train)
```

```python
# Creating a pickle file
filename = 'lr_model.pkl'
pickle.dump(Lr, open(filename, 'wb'))
```
